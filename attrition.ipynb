```python
# Fit and transform the OneHotEncoder using the categorical variable list
ohe = OneHotEncoder(sparse_output=False)
y_df_enc = pd.DataFrame(ohe.fit_transform(y_df[['Attrition','Department']]))
```

```python
y_df_enc
```

```python
#Assign Column names to the new encoded dataframe
y_df_enc.columns = ohe.get_feature_names_out(['Attrition','Department'])
y_df_enc
```

```python
#Split data in train and test
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df_enc, random_state=78)
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler = scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data

# The one hot coding was applied prior to the split, so not required to do again. Please see above
y_df_department_cols = ['Department_Human Resources','Department_Research & Development','Department_Sales']
y_df_department_cols
```

```python
#Check for Target Department values for training
y_train_department=y_train[y_df_department_cols].copy()
y_train_department.values
```

```python
#Check for Target Department values for testing
y_test_department=y_test[y_df_department_cols].copy()
y_test_department
```

```python
# Create a OneHotEncoder for the Attrition column
# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data

# The one hot coding was applied prior to the split, so not required to do again. Please see above
y_df_attrition_cols = ['Attrition_No','Attrition_Yes']
y_df_attrition_cols
```

```python
#Check for target attrition values in training dataset
y_train_attrition=y_train[y_df_attrition_cols].copy()
y_train_attrition.values
```

```python
#Check for target attrition values in testing dataset
y_test_attrition=y_test[y_df_attrition_cols].copy()
y_test_attrition
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
number_input_features = X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(X_train.shape[1],), name='input_features')

# Create at least two shared layers
shared_layer1 = layers.Dense(64, activation='relu', name='shared1')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu',name='shared2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_dense = layers.Dense(32, activation='relu',name='department_hidden')(shared_layer2)

# Create the output layer
department_output = layers.Dense(3, activation='softmax', name='department_output')(department_dense)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_dense = layers.Dense(32, activation='relu',name='attrition_hidden')(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(2, activation='sigmoid', name='attrition_output')(attrition_dense)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'department_output': 'categorical_crossentropy', 'attrition_output': 'binary_crossentropy'},
              metrics={'department_output': 'accuracy', 'attrition_output': 'accuracy'})

# Display the model summary
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {'department_output': y_train_department, 'attrition_output': y_train_attrition},
    epochs=100,
    batch_size=32,
    validation_split=0.2
)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test_scaled, {'department_output': y_test_department, 'attrition_output': y_test_attrition})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {test_results[3]}")
print(f"Attrition Accuracy: {test_results[4]}")
```

1. Accuracy is a commonly used metric to evaluate the performance of classification models, including Neural Networks. However, depending on the specific task and the characteristics of the data, accuracy might not always be the best metric to use, especially for branching models. Here's why:

  **Class Imbalance**: If the dataset is imbalanced, where some classes have significantly more samples than others, accuracy can be misleading. In this example attrition has significantly higher "No" than "Yes"

  **Model Complexity**: Branching models often aim to address multi-task learning or hierarchical feature extraction. In such cases, different branches of the model might have different objectives or focus on different aspects of the data. In such cases, optimizing solely for accuracy might not align with the overall goals of the model. Each branch might have its own set of evaluation metrics based on its specific task or objectives.
  
  **Misclassification Costs**: In certain applications, misclassifying one class might be more costly than misclassifying another. Accuracy treats all misclassifications equally, which might not reflect the true impact of model errors. For instance, in medical diagnosis, mistakenly identifying a healthy patient as diseased (false positive) might have different consequences compared to missing a diseased patient (false negative). In such scenarios, metrics like sensitivity (recall), specificity, or ROC-AUC might be more informative.

2. When choosing activation functions for the output layers of Neural Networks models, the selection depends on the nature of the task, the type of problem being solved, and the desired properties of the network's output. Here are some activation functions for output layers for this exercise:

  **Softmax Activation**: Softmax is commonly used in multi-class classification problems where the network needs to output probabilities for each class.  

  **Sigmoid Activation**: Sigmoid activation function is often used in binary classification problems where the network needs to predict a binary outcome (0 or 1).

  **ReLU Activation**: ReLU (Rectified Linear Unit) is a widely used activation function in hidden layers due to its simplicity and effectiveness. However, ReLU is not directly suitable for output layers in classification tasks because it doesn't bound the output values between specific ranges.
